## By statistical method -IQR

```python
# Select all numeric columns
num_cols = df.select_dtypes(include=['int64', 'float64']).columns.tolist()

# Loop through each numeric column and count outliers
for col in num_cols:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    outliers = df[(df[col] < lower_bound) | (df[col] > upper_bound)]
    print(f"{col}: {len(outliers)} outliers")
```

## Handling outliers

```python
import numpy as np

for col in ['Age', 'SibSp', 'Parch', 'Fare']:
    # Calculate bounds
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    
    # Count outliers before handling
    before_outliers = ((df[col] < lower_bound) | (df[col] > upper_bound)).sum()
    
    # Cap the outliers
    df[col] = np.where(df[col] > upper_bound, upper_bound,
                       np.where(df[col] < lower_bound, lower_bound, df[col]))
    
    # Count outliers after handling
    after_outliers = ((df[col] < lower_bound) | (df[col] > upper_bound)).sum()
    
    print(f"{col}: Before handling = {before_outliers}, After handling = {after_outliers}")
```

# Visuals

 ## Pairplot â€“ Relationships & Trends

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Select relevant columns for pairplot
sns.pairplot(df[['Age', 'Fare', 'Pclass', 'Survived']], hue='Survived', diag_kind='kde')
plt.suptitle("Pairplot: Relationships Between Key Features", y=1.02)
plt.show()

```

```python

# Observation:
# - Survived passengers generally had higher fares and were often in higher classes.
# - Age distribution shows no strong separation between survivors and non-survivors.
```

 ## Heatmap â€“ Correlation Matrix


```python


corr = df.select_dtypes(include=['int64', 'float64']).corr()

plt.figure(figsize=(8, 6))
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Correlation Heatmap")
plt.show()
```

## Histograms â€“ Distribution of Numeric Columns

```python
num_cols = df.select_dtypes(include=['int64', 'float64']).columns

df[num_cols].hist(bins=20, figsize=(12, 10), color='skyblue', edgecolor='black')
plt.suptitle("Histograms of Numeric Features", y=1.02)
plt.show()
```

```python
# Observation:
# - Most passengers paid low fares; only a few paid high fares.
# - Majority were young to middle-aged, with some children and elderly passengers.
```

## Scatterplots â€“ Key Relationships

```python
# Age vs Fare colored by survival
plt.figure(figsize=(8, 6))
sns.scatterplot(x='Age', y='Fare', hue='Survived', data=df, palette='Set1')
plt.title("Age vs Fare by Survival")
plt.show()
```

## Countplots â€“ Categorical Insights

```python
# Survival by Sex
sns.countplot(x='Sex', hue='Survived', data=df, palette='pastel')
plt.title("Survival Count by Gender")
plt.show()

# Observation:
# - Females had a higher survival rate than males.

# Survival by Pclass
sns.countplot(x='Pclass', hue='Survived', data=df, palette='muted')
plt.title("Survival Count by Passenger Class")
plt.show()
```
